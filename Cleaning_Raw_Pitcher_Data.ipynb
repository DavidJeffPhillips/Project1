```python
import pandas as pd
import cpi
```

```python
pitch_war_raw_df = pd.read_csv("data/war_daily_pitch.csv")
pitch_war_raw_df.head()
```

```python
PrelimPitch_df = pitch_war_raw_df[["name_common","age","mlb_ID", "player_ID","year_ID","stint_ID","team_ID","lg_ID","salary","ERA_plus","WAR"]]
PrelimPitch_df.head()
```

```python
PrelimPitch_df = PrelimPitch_df[PrelimPitch_df["year_ID"]>1984]
PrelimPitch_df = PrelimPitch_df[PrelimPitch_df["year_ID"]<2016]
PrelimPitch_df
```

```python
CleanPitch_df = PrelimPitch_df[PrelimPitch_df['salary']>10]
CleanPitch_df['sumWAR'] = PrelimPitch_df.groupby(by=['player_ID', 'year_ID'])['WAR'].transform('sum')
CleanPitch_df['avgERAplus'] = PrelimPitch_df.groupby(by=['player_ID','year_ID'])['ERA_plus'].transform('mean')
CleanPitch_df
```

```python
CleanPitch_df = CleanPitch_df.dropna()
CleanPitch_df
```

```python
salary_adj = CleanPitch_df.apply(lambda x: cpi.inflate(x['salary'], x['year_ID'], to=2015), axis=1)
salary_adj
CleanPitch_df['adjusted_sal'] = salary_adj
CleanPitch_df
```

```python
CleanPitch_df["team_ID"] = CleanPitch_df["team_ID"].replace(
    {"CAL": "LAA", "ANA": "LAA", "MON": "WSN"})
CleanPitch_df
```

```python
CleanPitch_df.to_csv("data/Clean_Pitch.csv", index=False)
```
