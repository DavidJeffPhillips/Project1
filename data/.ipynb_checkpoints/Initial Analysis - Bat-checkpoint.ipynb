```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batter_df = pd.read_csv("Clean_Bat.csv")
batter_df.head()
```

```python
x_axis = batter_df['salary']
war = batter_df['WAR']
plt.xlim(0, 30000000)
plt.ylim(-7,15)
plt.title("Salary vs WAR, 1976-2016")
plt.xlabel("Salary")
plt.ylabel("WAR")
plt.grid()

plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25)

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
batter_76_df = batter_df.loc[(
    batter_df["year_ID"] == 1976)]
batter_76_df.head()
```

```python
x_axis = batter_76_df['salary']
war = batter_76_df['WAR']
plt.xlim(0, 300000)
plt.ylim(-7,15)
plt.title("Salary vs WAR, 1976")
plt.xlabel("Salary")
plt.ylabel("WAR")
plt.grid()

plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25)

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
batter_96_df = batter_df.loc[(
    batter_df["year_ID"] == 1996)]
batter_96_df.head()
```

```python
x_axis = batter_96_df['salary']
war = batter_96_df['WAR']
plt.xlim(0, 10000000)
plt.ylim(-7,15)
plt.title("Salary vs WAR, 1996")
plt.xlabel("Salary")
plt.ylabel("WAR")
plt.grid()

plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25)

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
batter_16_df = batter_df.loc[(
    batter_df["year_ID"] == 2016)]
batter_16_df.head()
```

```python
x_axis = batter_16_df['salary']
war = batter_16_df['WAR']
plt.xlim(0, 30000000)
plt.ylim(-7,15)
plt.title("Salary vs WAR, 2016")
plt.xlabel("Salary")
plt.ylabel("WAR")
plt.grid()

plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25)

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
batter_yearly_df = batter_df.groupby("year_ID")
avg_yearly_war = batter_yearly_df["WAR"].mean()
avg_yearly_war_df = pd.DataFrame(avg_yearly_war)
avg_yearly_salary = batter_yearly_df['salary'].mean()
avg_yearly_salary_df = pd.DataFrame(avg_yearly_salary)
yearly_avg_df = pd.merge(avg_yearly_war_df, avg_yearly_salary_df, on="year_ID")
yearly_avg_df.head()
```

```python
x_axis = yearly_avg_df['salary']
war = yearly_avg_df['WAR']
plt.xlim(0, 1000000)
plt.ylim(-0,2.5)
plt.title("Avg Salary vs Avg WAR, 1976-2016")
plt.xlabel("Salary")
plt.ylabel("WAR")
plt.grid()

plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25)

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
t = yearly_avg_df.index
data1 = yearly_avg_df['WAR']
data2 = yearly_avg_df['salary']

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Year')
ax1.set_ylabel('WAR', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Salary', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)
```

```python
team_df = batter_df.groupby("team_ID")
team_war = team_df["WAR"].mean()
team_war_df = pd.DataFrame(team_war)
team_war_df
team_salary = team_df['salary'].mean()
team_salary_df = pd.DataFrame(team_salary)
combined_team_df = pd.merge(team_war_df, team_salary_df, on='team_ID')
combined_team_df.head()
```

```python
x_axis = combined_team_df['salary']
war = combined_team_df['WAR']
plt.xlim(0, 5000000)
plt.ylim(-0,2.5)
plt.title("Avg Team Salary vs Avg Team WAR, 1976-2016")
plt.xlabel("Salary")
plt.ylabel("WAR")
plt.grid()

plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25)

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
t = combined_team_df.index
data1 = combined_team_df['WAR']
data2 = combined_team_df['salary']

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Team')
ax1.set_ylabel('WAR', color=color)
ax1.bar(t, data1, color=color, alpha=0.5, align="center")
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Salary', color=color)  # we already handled the x-label with ax1
ax2.bar(t, data2, color=color, alpha=0.5, align="center")
ax2.tick_params(axis='y', labelcolor=color)
```
