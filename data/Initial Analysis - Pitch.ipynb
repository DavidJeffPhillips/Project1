```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
pitcher_df = pd.read_csv("Clean_Pitch.csv")
pitcher_df.head()
```

```python
x_axis = pitcher_df['adjusted_sal']
war = pitcher_df['WAR']
plt.xlim(0, 30000000)
plt.ylim(-5,13)
plt.title("Salary vs WAR, 1986-2016")
plt.xlabel("Inflation Adjusted Salary (2016 $)")
plt.ylabel("WAR")
plt.grid()

z = np.polyfit(x_axis, war, 1)
p = np.poly1d(z)
plt.plot(x_axis,p(x_axis),"r--")


plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25)

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
pitcher_yearly_df = pitcher_df.groupby("year_ID")
avg_yearly_war = pitcher_yearly_df["sumWAR"].mean()
avg_yearly_war_df = pd.DataFrame(avg_yearly_war)
avg_yearly_salary = pitcher_yearly_df['adjusted_sal'].mean()
avg_yearly_salary_df = pd.DataFrame(avg_yearly_salary)
yearly_avg_df = pd.merge(avg_yearly_war_df, avg_yearly_salary_df, on="year_ID")
yearly_avg_df.head()
```

```python
cost_war = yearly_avg_df["adjusted_sal"]/yearly_avg_df['sumWAR']
cost_war
yearly_avg_df['sal/war'] = cost_war
yearly_avg_df.head()
```

```python
x_axis = yearly_avg_df['adjusted_sal']
war = yearly_avg_df['sumWAR']
plt.xlim(0, 6000000)
plt.ylim(0.5,2.5)
plt.title("Avg Salary vs Avg WAR, 1986-2016")
plt.xlabel("Inflation Adjusted Salary (2016 $)")
plt.ylabel("WAR")
plt.grid()

z = np.polyfit(x_axis, war, 1)
p = np.poly1d(z)
plt.plot(x_axis,p(x_axis),"r--")

plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25)

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
t = yearly_avg_df.index
data1 = yearly_avg_df['sumWAR']
data2 = yearly_avg_df['adjusted_sal']

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Year')
ax1.set_ylabel('Avg WAR', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Avg Inflation Adjusted Salary (2015 $)', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)
```

```python
plt.title("Avg Cost per WAR per Year")
plt.xlabel("Year")
plt.ylabel("Avg Cost per WAR")
plt.grid()
plt.plot(yearly_avg_df['sal/war'], color="red")
```

```python
team_df = pitcher_df.groupby("team_ID")
team_war = team_df["sumWAR"].mean()
team_war_df = pd.DataFrame(team_war)
team_war_df
team_salary = team_df['adjusted_sal'].mean()
team_salary_df = pd.DataFrame(team_salary)
combined_team_df = pd.merge(team_war_df, team_salary_df, on='team_ID')
combined_team_df.head()
```

```python
x_axis = combined_team_df['adjusted_sal']
war = combined_team_df['sumWAR']
plt.xlim(1000000, 4000000)
plt.ylim(0.5,2.0)
plt.title("Avg Team Salary vs Avg Team WAR, 1985-2015")
plt.xlabel("Avg Inflation Adjusted Salary")
plt.ylabel("Avg WAR")
plt.grid()

z = np.polyfit(x_axis, war, 1)
p = np.poly1d(z)
plt.plot(x_axis,p(x_axis),"r--")

plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25)

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
t = combined_team_df.index
data1 = combined_team_df['sumWAR']
data2 = combined_team_df['adjusted_sal']

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Team')
ax1.set_ylabel('Avg WAR', color=color)
ax1.bar(t, data1, color=color, alpha=0.5, align="center")
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Avg Inflation Adjusted Salary(2015 $)', color=color)  # we already handled the x-label with ax1
ax2.bar(t, data2, color=color, alpha=0.5, align="center")
ax2.tick_params(axis='y', labelcolor=color)
```

```python
team_df = pitcher_df.groupby(["team_ID", "year_ID"])
team_df.head()
yearly_team_war = team_df['sumWAR'].sum()
yearly_team_war
yearly_team_sal = team_df['adjusted_sal'].sum()
yearly_team_sal
yearly_costof_war = yearly_team_sal/yearly_team_war
yearly_costof_war
yearly_sal_war_df = pd.DataFrame({"Yearly War by Team": yearly_team_war,
                           "Yearly Salary by Team": yearly_team_sal,
                           "Yearly cost of WAR by Team": yearly_costof_war})
yearly_sal_war_df.head()

new = yearly_sal_war_df["Yearly cost of WAR by Team"]
new_df = pd.DataFrame(new)
new_df.head()
```
