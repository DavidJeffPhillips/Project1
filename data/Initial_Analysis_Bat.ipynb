```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
batter_df = pd.read_csv("Clean_Bat.csv")
batter_df.head()
```

```python
x_axis = batter_df['adjusted_sal']
war = batter_df['sumWAR']
plt.xlim(0, 30000000)
plt.ylim(-5,13)
plt.title("Salary vs WAR, 1986-2016")
plt.xlabel("Inflation Adjusted Salary (2016 $)")
plt.ylabel("WAR")
plt.grid()

z = np.polyfit(x_axis, war, 1)
p = np.poly1d(z)
plt.plot(x_axis,p(x_axis),"r--")
plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25, edgecolor='black', color='blue')

#plt.savefig("salary_vs_war.png")

plt.show()
correlation8616 = np.corrcoef(x_axis, war) [0,1]
correlation8616 = correlation8616**2
print(correlation8616)
```

```python
batter_yearly_df = batter_df.groupby("year_ID")
avg_yearly_war = batter_yearly_df["sumWAR"].mean()
avg_yearly_war_df = pd.DataFrame(avg_yearly_war)
avg_yearly_salary = batter_yearly_df['adjusted_sal'].mean()
avg_yearly_salary_df = pd.DataFrame(avg_yearly_salary)
yearly_avg_df = pd.merge(avg_yearly_war_df, avg_yearly_salary_df, on="year_ID")
yearly_avg_df.head()
```

```python
cost_war = yearly_avg_df["adjusted_sal"]/yearly_avg_df['sumWAR']
cost_war
yearly_avg_df['sal/war'] = cost_war
yearly_avg_df.head()
```

```python
x_axis = yearly_avg_df['adjusted_sal']
war = yearly_avg_df['sumWAR']
plt.xlim(0, 6000000)
plt.ylim(0.5,2.5)
plt.title("Avg Salary vs Avg WAR, 1986-2016")
plt.xlabel("Inflation Adjusted Salary (2016 $)")
plt.ylabel("WAR")
plt.grid()

z = np.polyfit(x_axis, war, 1)
p = np.poly1d(z)
plt.plot(x_axis,p(x_axis),"r--")
plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25,edgecolor='black', color='blue')

plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25,edgecolor='black', color='blue')

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
t = yearly_avg_df.index
data1 = yearly_avg_df['sumWAR']
data2 = yearly_avg_df['adjusted_sal']

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Year')
ax1.set_ylabel('Avg WAR', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Avg Inflation Adjusted Salary (2016 $)', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)
```

```python
batter_df.set_index(['team_ID', 'year_ID'])
tteam_df = batter_df.groupby(["team_ID"])
tteam_df.head()
team_war = tteam_df["sumWAR"].mean()
team_war_df = pd.DataFrame(team_war)
team_war_df
team_salary = tteam_df['adjusted_sal'].mean()
team_salary_df = pd.DataFrame(team_salary)
combined_team_df = pd.merge(team_war_df, team_salary_df, on='team_ID')
combined_team_df.head()
```

```python
x_axis = combined_team_df['adjusted_sal']
war = combined_team_df['sumWAR']
plt.xlim(1000000, 4000000)
plt.ylim(0.5,2.0)
plt.title("Avg Team Salary vs Avg Team WAR, 1986-2016")
plt.xlabel("Avg Inflation Adjusted Salary")
plt.ylabel("Avg WAR")
plt.grid()

z = np.polyfit(x_axis, war, 1)
p = np.poly1d(z)
plt.plot(x_axis,p(x_axis),"r--")
plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25,edgecolor='black', color='blue')

plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25,edgecolor='black', color='blue')

#plt.savefig("salary_vs_war.png")

plt.show()
```

```python
t = combined_team_df.index
data1 = combined_team_df['sumWAR']
data2 = combined_team_df['adjusted_sal']

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Team')
ax1.set_ylabel('Avg WAR', color=color)
ax1.bar(t, data1, color=color, alpha=0.5, align="center")
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Avg Inflation Adjusted Salary(2016 $)', color=color)  # we already handled the x-label with ax1
ax2.bar(t, data2, color=color, alpha=0.5, align="center")
ax2.tick_params(axis='y', labelcolor=color)
```

```python
years = [1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016]
correlations = []
for year in years:
    batteryear_df = batter_df.loc[(batter_df["year_ID"] == year)]
    x_axis = batteryear_df['adjusted_sal']
    war = batteryear_df['sumWAR']
    z = np.polyfit(x_axis, war, 1)
    p = np.poly1d(z)
    plt.plot(x_axis,p(x_axis),"r--")
    plt.scatter(x_axis, war, marker="o", s=50, alpha=0.25,edgecolor='black', color='blue')
    correlationyear = np.corrcoef(x_axis, war) [0,1]
    correlationyear = correlationyear**2
    correlations.append(correlationyear)


correlations


# plt.title("Salary vs WAR, 1986-2016")
# plt.xlabel("Inflation Adjusted Salary (2016 $)")
# plt.ylabel("WAR")
# plt.grid()
#plt.savefig("salary_vs_war.png")
```

```python
plt.plot(years, correlations)
plt.title("Correlation between WAR and adjusted salary between 1986 and 2016")
plt.xlabel("Year")
plt.ylabel("R-squared of adjusted salary and WAR")
plt.grid()
```
