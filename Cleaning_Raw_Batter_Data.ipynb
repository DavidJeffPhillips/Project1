```python
import pandas as pd
import cpi
```

```python
BatRaw_csv = "data/war_daily_bat.csv"
```

```python
BatRaw_df = pd.read_csv(BatRaw_csv)
BatRaw_df.head()
```

```python
PrelimBat_df = BatRaw_df[["name_common","age","mlb_ID", "player_ID","year_ID","stint_ID","team_ID","lg_ID","pitcher","salary","OPS_plus","WAR"]]
PrelimBat_df
```

```python
PrelimBat_df=PrelimBat_df[PrelimBat_df["year_ID"]>1984]
PrelimBat_df=PrelimBat_df[PrelimBat_df["year_ID"]<2016]
PrelimBat_df=PrelimBat_df[PrelimBat_df["pitcher"]=="N"]
PrelimBat_df
```

```python
CleanBat_df = PrelimBat_df[PrelimBat_df['salary']>10]
CleanBat_df['sumWAR'] = PrelimBat_df.groupby(by=['player_ID', 'year_ID'])['WAR'].transform('sum')
CleanBat_df['avgOPSplus'] = PrelimBat_df.groupby(by=['player_ID','year_ID'])['OPS_plus'].transform('mean')
CleanBat_df
```

```python
CleanBat_df = CleanBat_df.dropna()
CleanBat_df
```

```python
salary_adj = CleanBat_df.apply(lambda x: cpi.inflate(x['salary'], x['year_ID'], to=2015), axis=1)
salary_adj
CleanBat_df['adjusted_sal'] = salary_adj
CleanBat_df
```

```python
CleanBat_df["team_ID"] = CleanBat_df["team_ID"].replace(
    {"CAL": "LAA", "ANA": "LAA", "MON": "WSN"})
CleanBat_df
```

```python
CleanBat_df.to_csv("data/Clean_Bat.csv", index=False)
```
